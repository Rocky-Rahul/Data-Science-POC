**Data Cleaning :**\
"There is no missing values in this dataframe.So, no need to data cleaning process."

```python
# clear irrelevant columns in the dataframe
df = df.drop(columns=['User ID'], axis=0)
```

```python
df
```

**Label Encoding :**\
It is a technique, used to convert categorical data into numerical format

```python
# import LebelEncoder library
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df
```

**X & Y Split:**


```python
# separate the independent variables in the dataframe
x = df.iloc[:,:-1]
x
```

```python
# separate the dependent variables in the dataframe
y = df.iloc[:,-1]
y
```

**Train & Test Split :**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=41)
```

**Normalization:**
Standard scalar

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x_train = std.fit_transform(x_train)
x_test = std.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

**Model Building_1 :**\
Decision Trees Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier()
dc.fit(x_train,y_train)
```

```python
y_predict = dc.predict(x_test)
y_predict
```

```python
# it shows the tree depth for this data
from sklearn import tree
plt.figure(figsize =(100,80))
tree.plot_tree(dc,filled = True)
```

**Evaluation Metrics:**\
Decision Tree

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
ac_1 = accuracy_score(y_predict,y_test)
ac_1
```

```python
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```

**Model Building_2:**\
Support Vector Macine

```python
from sklearn import svm
c = 1
svm_model = svm.SVC(kernel='poly', degree=3,C=c,gamma = 'auto')
svm_model.fit(x_train,y_train)
```

```python
y_pre = svm_model.predict(x_test)
y_pre
```

**Evaluation Metric:**\
SVM

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
ac_2 = accuracy_score(y_pre,y_test)
ac_2
```

```python
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```

**Model Building_3:**\
K-Neighbors Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_predict2 = knn.predict(x_test)
y_predict2
```

**Evaluation Metric:**\
Knn

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
# compare the three model we build
class_name = ("Decision Tree",'SVM',"KNN")
class_score = (ac_1,ac_2,ac_3)
y_pos= np.arange(len(class_score))
colors = ("red","gray","purple","green","orange")
plt.figure(figsize=(16,9))
plt.bar(y_pos,class_score,color=colors)
plt.xticks(y_pos,class_name,fontsize=20)
plt.ylabel('Accuracy')
plt.grid()
plt.title(" Confusion Matrix Comparision of the Classes",fontsize=15)
plt.show()
```
